# Ein erster minimaler Workflow
## 1. Imports und das Modell

```python
# pip install langgraph

from typing import TypedDict
from langgraph.graph import StateGraph, START, END

# TODO: hier das LLM einsetzen, mit dem man oben verbinden konnte
llm = llm_marvin
```

## 2. Der State

Der State ist das "Gedächtnis" des Graphen. Jeder Node liest daraus und
schreibt ein **partielles Update** zurück (ein dict mit nur den Feldern,
die er ändert). LangGraph merged das automatisch in den Gesamt-State.

Definiert ein `TypedDict` mit zwei Feldern:
- `thema`: str — das Recherche-Thema
- `antwort`: str — die Antwort des Modells

```python
class AgentState(TypedDict):
    thema: str
    antwort: str
```

## 3. Ein Node

Ein Node ist eine Python-Funktion mit der Signatur `(state) -> dict`.

Baut einen Node `node_erste_einschaetzung`, der:
1. einen Prompt baut, der nach einer kurzen (3 Sätze) Einschätzung zum `state['thema']` fragt,
2. das Modell aufruft,
3. `{"antwort": <Modellantwort>}` zurückgibt.

```python
def node_erste_einschaetzung(state: AgentState) -> dict:
    # Gibt eine erste, kurze Einschätzung zum Thema ab.
    
    prompt = (
        f"Gib eine kurze (3 Sätze) erste Einschätzung zum folgenden Thema:\n"
        f"{state['thema']}"
    )
    response = llm.invoke(prompt)
    return {'antwort': response.content}
```

## 4. Der Graph

Baut den Graphen:
1. `StateGraph(AgentState)` erstellen
2. Node `"erste_einschaetzung"` hinzufügen (`add_node`)
3. Kante `START -> "erste_einschaetzung"` hinzufügen (`add_edge`)
4. Kante `"erste_einschaetzung" -> END` hinzufügen
5. Mit `.compile()` zum ausführbaren Graphen machen

```python
graph_builder = StateGraph(AgentState)

# Ecken
graph_builder.add_node('erste_einschaetzung', node_erste_einschaetzung)

# Kanten
graph_builder.add_edge(START, 'erste_einschaetzung')
graph_builder.add_edge('erste_einschaetzung', END)

graph = graph_builder.compile()
```

```python
# pip install ipython

from IPython.display import Image, display

display(Image(graph.get_graph().draw_mermaid_png()))
```

## 5. Thema definieren und Workflow ausführen

Ruft `graph.invoke(...)` mit einem Start-State auf (Dict mit `thema` und leerer `antwort`).

```python
# TODO: Optional ein andere Thema definieren
thema = 'Herausforderungen für IT-Berater im Jahr 2027'

ergebnis = graph.invoke({
    'thema': thema,
    'antwort': ''
})
print(ergebnis['antwort'])
```
